```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
range_ = [0.0, 0.7]
noise = [0.0, 0.02]
```

```python
dataset_size = 500
x = np.random.uniform(range_[0], range_[1], dataset_size)
y = x + 0.3*np.sin(2*np.pi*(x+np.random.normal(noise[0], noise[1], dataset_size))) + \
    0.3 * np.sin(4 * np.pi * (x + np.random.normal(noise[0], noise[1], dataset_size)))+ \
    np.random.normal(noise[0], noise[1], dataset_size)
```

```python
x_true = np.linspace(-0.5, 1.2, 200)
y_true = x_true + 0.3*np.sin(2*np.pi*(x_true)) + \
    0.3 * np.sin(4 * np.pi * (x_true))
```

```python
plt.plot(x, y, '*')
plt.plot(x_true, y_true)
```
